```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
DB_USER = 'root'
DB_PASSWORD = 'root'
DB_HOST = '10.0.11.46'
DB_PORT = '3310'
DB_NAME = 'retail_db'

MYSQL_URL = f"mysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

engine = create_engine(MYSQL_URL)
```

```python
df_customers = pd.read_sql("SELECT * FROM customers",engine)
```

```python
def extract_table(table_name: str) -> pd.DataFrame:
    
    query = f"SELECT * FROM {table_name}"

    try:
        df = pd.read_sql(query, engine)
        print(f"La tabla {table_name} se ha leido exitosamente")
    except Exception as e:
        print(f"Error al leer la tabla {table_name}: {e}")

    return df
```

```python
df_customers = extract_table('customers')
df_orders = extract_table('orders')
df_order_items = extract_table('order_items')
df_products = extract_table('products')
df_categories = extract_table('categories')
df_departments = extract_table('departments')
```

```python
df_customers.head()
```

```python
df_customers.info()
```

```python
df_orders.info()
```

```python
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'], errors = 'coerce')
df_orders.head()
```

```python
df_customers['customer_full_name'] = (
    df_customers['customer_fname'].str.strip() + ' ' + df_customers['customer_lname'].str.strip()
)
df_customers['customer_full_name']
```

```python
# Joins

df = df_orders.merge(
    df_customers, left_on='order_customer_id', right_on= 'customer_id', how='left'
)

df.shape
```

```python
df = df.merge(
    df_order_items, left_on='order_id', right_on='order_item_order_id', how='inner'
)

df.shape
```

```python
df = df.merge(
    df_products, left_on='order_item_product_id', right_on='product_id', how='left'
)
```

```python
df = df.merge(
    df_categories, left_on='product_category_id', right_on='category_id', how='left'
)
```

```python
df = df.merge(
    df_departments, left_on='category_department_id', right_on='department_id', how='left'
)
```

```python
df.head()
```

```python
sales_detail = df[[
    'order_id',
    'order_date',
    'order_status',
    'customer_id',
    'customer_full_name',
    'customer_email',
    'customer_city',
    'customer_state',
    'product_id',
    'product_name',
    'category_name',
    'department_name',
    'order_item_quantity',
    'order_item_product_price',
    'order_item_subtotal'
]].copy()

sales_detail.head()
```

```python
sales_detail.rename(columns={
    'order_id': 'sale_id',
    'order_date': 'sale_date',
    'order_status': 'sale_status',
    'customer_full_name': 'customer_name',
    'order_item_quantity': 'quantity',
    'order_item_product_price': 'unit_price',
    'order_item_subtotal': 'subtotal'
}, inplace=True)

sales_detail.head()
```

```python
sales_detail['year'] = sales_detail['sale_date'].dt.year
sales_detail['month'] = sales_detail['sale_date'].dt.month
sales_detail['month_name'] = sales_detail['sale_date'].dt.month_name()

sales_detail[['sale_date', 'year', 'month', 'month_name']].head()
```

```python
sales_detail['customer_name'] = sales_detail['customer_name'].fillna('Cliente desconocido')
sales_detail['product_name'] = sales_detail['product_name'].fillna('Producto desconocido')
sales_detail['category_name'] = sales_detail['category_name'].fillna('Sin categoría')
sales_detail['department_name'] = sales_detail['department_name'].fillna('Sin departamento')

sales_detail.head()
```

```python
sales_detail = sales_detail.dropna(subset=['sale_id', 'sale_date', 'product_id'])
sales_detail.shape
```

```python
sales_detail.head()
```

```python
sales_detail.to_csv('sales_detail.csv', index=False, encoding = 'utf-8')
```
